```python
import numpy as np
import scipy.stats as stats

```

```python
n_control = 1000
n_test = 1000
successes_control = 200
successes_test = 250

conversion_rate_control = successes_control / n_control
conversion_rate_test = successes_test / n_test
se_control = np.sqrt(conversion_rate_control * (1 - conversion_rate_control) / n_control)
se_test = np.sqrt(conversion_rate_test * (1 - conversion_rate_test) / n_test)
z_score = (conversion_rate_test - conversion_rate_control) / np.sqrt(se_test**2 + se_control**2)
p_value = stats.norm.sf(abs(z_score)) #one-sided
print(f'z-score: {z_score}, p-value: {p_value}')
```

```python
if p_value < 0.05:
    print("The change is statistically significant.")
else:
    print("The change is not statistically significant.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labels = ['Control', 'Test']
conversion_rates = [conversion_rate_control, conversion_rate_test]
```

```python
plt.bar(labels, conversion_rates, color=['blue', 'green'])
plt.xlabel('Group')
plt.ylabel('Conversion Rate')
plt.title('Conversion Rates for Control and Test Groups')
plt.show()
```
